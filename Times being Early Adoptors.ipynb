```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('sum_up.csv')
```

```python
df.fillna(0, inplace=True)
# df.head()
```

```python
df['Total Times'] = df.iloc[:,2:].sum(axis=1)
df.head()
```

```python
for i in range(2,8): 
    print(df.iloc[:, 1:i].sum(axis=1, numeric_only=True).to_list()[21])
```

```python
# df.sort_values(by='Total Times',ascending=False)
```

```python
target_df = df.sort_values(by='Total Times',ascending=False)

def plot_df(df): 
    font = {'size': 18}

    plt.rc('font', **font)
    
    x = df.iloc[:,1]

    colour = ['#A6A6A6', '#1665E5', '#EF2D0F', '#EA1E7B', '#FEB250', '#2ACB0B']
    cum_award = np.zeros(df.shape[0])

    plt.figure(figsize=(20,15))
    for i in range(2,8): 
        plt.bar(x, df.iloc[:, i], color = colour[i-2], bottom = cum_award, label = df.columns[i])
        cum_award = df.iloc[:, 2:i].sum(axis=1, numeric_only=True).to_list()

    plt.title('Times being Early Adoptors')
    plt.xlabel('Companies')
    plt.ylabel('Times being Early Adoptors')
    plt.legend()
    plt.xticks(rotation=90)
    plt.show()
    
plot_df(target_df)
```
