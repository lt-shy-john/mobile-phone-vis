```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('sum_up.csv')
```

```python
df.fillna(0, inplace=True)
df.head()
```

```python

for i in range(2,8): 
    print(df.iloc[:, 1:i].sum(axis=1, numeric_only=True).to_list()[21])
```

```python
x = df.iloc[:,1]

colour = ['#0F65EF', '#3ECC7D', '#EF2D0F', '#423ECC', '#FEB250', '#40BF38']
cum_award = np.zeros(df.shape[0])

plt.figure(figsize=(20,15))
for i in range(2,8): 
    plt.bar(x, df.iloc[:, i], color = colour[i-2], bottom = cum_award, label = df.columns[i])
    cum_award = df.iloc[:, 2:i].sum(axis=1, numeric_only=True).to_list()
    
plt.title('Times being Early Adoptors')
plt.xlabel('Companies')
plt.ylabel('Times being Early Adoptors')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
